Importing Modules

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt 
import os 
```

Loading the model and test images

```python
model = YOLO("best.pt")

image_path = "valid/images/pic-114-_jpg.rf.a0f30e06b3b96d7879d5f55a7012433c.jpg" 


if not os.path.exists(image_path):
    print(f"Error: The file path is wrong.\nI tried to look at: {os.path.abspath(image_path)}")
else:
    print("File found! Loading image...")
    
   
    img = cv2.imread(image_path)
    
    
    if img is None:
        print("Error: Path exists, but OpenCV failed to read it.")
    else:
        
        print(f"Image Size: {img.shape}")
        results = model.predict(img)
```

Printing the image and validating

```python
result_img = results[0].plot()
result_img_rgb = cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB)

plt.imshow(result_img_rgb)
plt.axis('off')
plt.show()
```
