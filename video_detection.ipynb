Importing the dependencies

```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("models/best.pt") 
cap = cv2.VideoCapture("sample_video.mp4")

# Get Video Properties 
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = int(cap.get(cv2.CAP_PROP_FPS))

# Create the "Writer" 
out = cv2.VideoWriter('demo_v2_production.mp4', 
                      cv2.VideoWriter_fourcc(*'mp4v'), 
                      fps, 
                      (width, height))

print("Processing video...")

while cap.isOpened():
    success, frame = cap.read() # success is a boolean which returns true/false based on if a frame was received or not 
    
    if not success:
        break 

    
    results = model.predict(frame, conf=0.25, verbose=False) # verbose keep notebook logs clean

    
    annotated_frame = results[0].plot()

    
    out.write(annotated_frame)


cap.release() 
out.release() # save the file to disk 
cv2.destroyAllWindows()

print("Done!")
```
